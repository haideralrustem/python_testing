```python
def player_defends(enemy_list, player):
    nods = player.defense_roll()
    print('\nnods = ', nods)
    i = len(enemy_list)
    e = enemy_list[0]
    swipes = e.boss_swipes(4)
    s_out = ''
    difference = 0
    if nods < swipes:
        difference = swipes - nods
    for swipe in range(0, difference) :
        if nods < swipes:
            d = e.atk - player.dmg_reduction
            if d >= 0:
                player.takes_damage(d) 
                s_out += 'you took : %s dmg\t\t '%d
        else:
            print('You dodgded this Swipe !!')
        i -= 1
    print(s_out, '\n')
    return enemy_list, player
    
```

```python
def show_bars(actor):
    real_max_hp = actor.max_hp
    real_current_hp = actor.hp
    max_hp_rescaled = int(real_max_hp * 1.0)  # we are rescaling by 80%. You can change anytime according to your view
    current_hp_rescaled = int(real_current_hp * 1.0)
    health = "[ "
    for i in range(0, max_hp_rescaled):
        if i <= current_hp_rescaled:
            health += "|"
        else:
            health += ' '
    return health + ' ] \t\t'
```

```python
def combat_flow(enemy_list, dead_enemies, player):
    
    n = len(enemy_list)
    c = 0
    show_deaths = ''
    if n > 0 and player.hp > 0:
        print(show_bars(player))
        player.show_player_mana()
        print(player.hp)
        show_head = ""
        show_leg = ""
        total_bars = ""
        show_stats = ""
        for e in enemy_list:
            show_head += "\tO \t\t\t"  
            show_leg += "\t/|\\ \t\t\t"
            total_bars +=  show_bars(e)
            show_stats += "%s \t\t"%(e)
        print(show_head)
        print(show_leg)
        print(total_bars)
        print(show_stats , '\n')
        #    --------------- when boss is True -----------------
        if (enemy_list[0].boss == True):
            action = randint(0, 5)
            if action in [3, 4, 5]:
                print(' normal action ')
                pass
            elif action in [1, 2]:
                print('Boss starts to swipe....')
                
                ##................. 
                enemy_list, player = player_defends(enemy_list, player)
                print('after player defends:')
                
            else:
                print('\t\tACTION SEQUENCE INCOMING 3 !!')
                return
        ab, abtext = player.player_roll()
        print('\n---->  ', ab.ab_icon, '\t\t', abtext.upper(), ' <------')
        next_enemy = enemy_list[n-1]
        #------------------ AOE---------------
        if abtext == 'blade storm':
            for en in enemy_list:
                dmg= ab(en, player)
                if en.is_dead():
                    c += 1
            while(c > 0):
                
                dead = enemy_list.pop()
                dead_enemies.insert(0, dead)
                c -= 1
            print('Damage dealt (AOE) --> ', dmg)
        
        #---------------------------------------------------         
        else:
            dmg = ab(next_enemy, player)
            print('Damage dealt --> ', dmg)
            if next_enemy.is_dead() and n > 0:
                dead = enemy_list.pop()
                dead_enemies.insert(0, dead)
                show_head2 = ""
                show_leg2 = ""
                show_stats2 = ""
                
                show_head2 += "\tO \t\t\t"  
                show_leg2 += "\t/|\\ \t\t\t"
               
                show_stats2 += "%s \t\t"%(dead)
                print(show_head2)
                print(show_leg2)
                print(show_stats2)
        print(' ........................ ............................. \n', 'enemies left:', len(enemy_list))
        show_head = ""
        show_leg = ""
        show_stats = ""
        total_bars = ""
        for e in enemy_list:
            show_head += "\tO \t\t\t"  
            show_leg += "\t/|\\ \t\t\t"
            total_bars += show_bars(e)
            show_stats += "%s \t\t"%(e)
        if len(enemy_list) == 0:
            print('\n >-->o')
        else:
            print('\n')
            print(show_head)
            print(show_leg)
            print(total_bars)
            print(show_stats)
        nods = player.defense_roll()
        print('\nnods = ', nods)
        i = len(enemy_list)
        s_out = ''
        for e in enemy_list:
            if i > nods:
                d = e.atk - player.dmg_reduction
                if d >= 0:
                    # player taking damage
                    player.takes_damage(d) 
                    s_out += 'you took : %s dmg\t\t '%d
            i -= 1
        print(s_out)
        print(show_bars(player))
        player.show_player_mana()
#         for e in enemy_list():

    else:
        if player.hp > 0:
            print('All enemies DEAD!!\n')
            for e in dead_enemies:
                show_deaths += '(x_x) \t\t'
            print(show_deaths)
        else:
            print('\t YOU\tDIED\t')
    return dead_enemies
```
