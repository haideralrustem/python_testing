```python
def player_defends(enemy_list, player):
    nods = player.defense_roll()
    print('\nnods = ', nods)
    i = len(enemy_list)
    e = enemy_list[0]
    swipes = e.boss_swipes(4)
    s_out = ''
    difference = 0
    if nods < swipes:
        difference = swipes - nods
    for swipe in range(0, difference) :
        if nods < swipes:
            d = e.atk - player.dmg_reduction
            if d >= 0:
                player.takes_damage(d) 
                s_out += 'you took : %s dmg\t\t '%d
        else:
            print('You dodgded this Swipe !!')
        i -= 1
    print(s_out, '\n')
    return enemy_list, player
    
```

```python
def show_bars(actor):
    real_max_hp = actor.max_hp
    real_current_hp = actor.hp
    max_hp_rescaled = int(real_max_hp * 1.0)  # we are rescaling by 80%. You can change anytime according to your view
    current_hp_rescaled = int(real_current_hp * 1.0)
    health = "📎"
    for i in range(0, max_hp_rescaled):
        if i <= current_hp_rescaled:
            health += "|"
        else:
            health += ' '
    return  " {}📎 {} \t\t".format(health, actor.hp)
```

```python
def player_steals_life(player, dmg):
    heal_amount = dmg * player.lifesteal
    player.hp += heal_amount
    if (player.hp > player.max_hp):
        player.hp = player.max_hp
```

```python
def combat_flow(enemy_list, dead_enemies, player):
    
    bleed_value = 0
    if(player.style == "zed"):      # <-------------------------------------  Character Zed bleed value buff
        bleed_value = bleed_value * 2    
    n = len(enemy_list)
    c = 0
    show_deaths = ''
    turn_number = 0
    if n > 0 and player.hp > 0:
        # ....... Show Health and Stats.................
        print(show_bars(player))
        player.show_player_mana()
        print()
        show_head = ""
        total_bars = ""
        show_stats = ""
        for e in enemy_list:
            show_head += "\t 🧛  \t\t\t\t\t" 
            total_bars +=  show_bars(e)
            show_stats += "%s \t\t\t"%(e)
        print(show_head)
        print(total_bars)
        print(show_stats , '\n')
        #    --------------- when boss is True -----------------
        if (enemy_list[0].boss == True):
            action = randint(0, 5)
            if action in [3, 4, 5]:
                print(' normal action ')
                pass
            elif action in [1, 2]:
                print('Boss starts to swipe....')
                
                ##................. 
                enemy_list, player = player_defends(enemy_list, player)
                print('after player defends:')
                
            else:
                print('\t\tACTION SEQUENCE INCOMING 3 !!')
                return
        
        # .......... Battle flow continues by player selecting an ability.........................
        
        ab, abtext = player.player_roll()   #<------ Ability is generated
        print('\n---->  ', ab.ab_icon, '\t\t', abtext.upper(), ' <------')
        next_enemy = enemy_list[n-1]
        
        if(next_enemy.hp <= 0):  # <-------- Testing if enemy reached 0 before player dmg is applied
            dead = enemy_list.pop()
            dead_enemies.insert(0, dead)
            if(len(enemy_list) > 0):
                next_enemy = enemy_list[(len(enemy_list)) - 1]  # update next_enemy 
        
        if abtext in status_effect_abilities:
            if abtext == "back stab":
                next_enemy.bleed = True  #   <-------------   Tagging enemy as bleeding
        
        #------------------ AOE  damage only ---------------------------------------------------------
        if abtext == 'blade storm':
            for en in enemy_list:
                dmg= ab(en, player)  # -->  this line inflicts the damage
                player_steals_life(player, dmg)  # <---- Lifesteal is applied
                
                # ..... Testing for bleed  (bleed is removed upon death).........
                if en.bleed == True:
                    bleed_value += Ability.back_stab.bleed_value
                    en.hp -= bleed_value       # <---------  bleed dmg registers
                    print('enemy bleeds for ', '+ ', bleed_value, ' ❣️❣️')
                if en.is_dead():
                    c += 1
                    en.bleed = False    # <---------------   Remove bleed
                    
            while(c > 0):
                dead = enemy_list.pop()
                dead_enemies.insert(0, dead)
                c -= 1
            if en.bleed == True:
                print('Damage dealt --> ', dmg, ' + ', bleed_value)  # <---- if bleed true
            else:
                print('Damage dealt (AOE) --> ', dmg)
        
        #------------------------------------------------------------------------------------------------------     
        
        
        
        else:  # .....Regular abilities damage calculation......
            
            dmg = ab(next_enemy, player)  #  Damage is inflicted..
            player_steals_life(player, dmg)   # <---- Lifesteal is applied
            
            # test for additional damage such as bleed or poison
            if next_enemy.bleed == True:
                bleed_value += Ability.back_stab.bleed_value
                next_enemy.hp -= bleed_value
                print('enemy bleeds for ', '+ ', bleed_value, ' ❣️❣️')          
                print('Damage dealt --> ', dmg, ' + ', bleed_value)
            else:
                print('Damage dealt --> ', dmg)  # <------ print dmg without bleed
                
            if next_enemy.is_dead() and n > 0:
                next_enemy.bleed = False   # <---------------   Remove bleed
                dead = enemy_list.pop()
                dead_enemies.insert(0, dead)
                show_head2 = ""
                show_leg2 = ""
                show_stats2 = ""
                
                show_head2 += "\tO \t\t\t\t"  
                show_leg2 += "\t/|\\ \t\t\t\t"
               
                show_stats2 += "%s \t\t\t"%(dead)
                print(show_head2)
                print(show_leg2)
                print(show_stats2)
        print(' ........................ ............................. \n', 'enemies left:', len(enemy_list))
        show_head = ""
        show_stats = ""
        total_bars = ""
        for e in enemy_list:
            show_head += "\t🧛  \t\t\t\t\t\t"  
            total_bars += show_bars(e)
            show_stats += "%s \t\t\t"%(e)
        if len(enemy_list) == 0:
            print('\n >-->o')
        else:
            print('\n')
            print(show_head)
            print(total_bars)
            print(show_stats)
            
        nods = player.defense_roll()
        print('\nnods = ', nods)
        i = len(enemy_list)
        s_out = ''
        
        # Enemy Attacks / Player Defends
        
        #    ---   Player  DEFENSE stage (for loop takes in each enemy's dmg)   ---- 
        
        # ...... Add any damage reducing abilities/spells here (variable d).......
        for e in enemy_list: 
            if i > nods:
                d = e.atk - player.dmg_reduction
                if(player.last_spell in ['energy barrier']):  #  <----------- dmg reduction modified 
                        barrier_reduction = Magic.energy_barrier.damage_reduction_bonus
                        if(player.style=="paladin"):
                            barrier_reduction = barrier_reduction*1.5
                            player.mana += 1
                        d = e.atk - (player.dmg_reduction + barrier_reduction)
                        print("your shield reduces extra dmg!! 💎💎💎")                        
                
                if d >= 0:
                    if(e.stunned == True):  # <------------ Testing for 💫 stunned state
                        print('enemy 💫 did not attack')
                        e.stunned == False
                        continue
                        
                    player.takes_damage(d)    #    player taking damage  
                    s_out += 'you took : %s dmg\t\t '%d
            i -= 1
        
        # reset palyer.last_spell so that this spell works for one turn only
        player.last_spell = 'Empty'
        
        print(s_out)
        print(show_bars(player))
        player.show_player_mana()

    else:
        if player.hp > 0:
            print('All enemies DEAD!!\n')
            for e in dead_enemies:
                show_deaths += '(x_x) \t\t'
            print(show_deaths)
        else:
            print('\t YOU\tDIED\t')
    return dead_enemies
```
